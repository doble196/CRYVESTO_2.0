```python
df_coin= ld.load_from_yahoo(['BTC-USD'], start,end)
```

```python
df_coin.index
```

```python
joined_df = sentiment_df.join(df_coin['BTC-USD']['close'])
```

```python
joined_df['daily_ret']=joined_df['close'].pct_change().shift()
```

```python
joined_df
```

```python
#polarity=[]
#subj=[]
signal=[]
#for i in f_df['Headline']:
for i in joined_df.index:
    #Bear_sum	Bull_sum	scores	
    if (joined_df['Bull_sum'][i] > 700) and (joined_df['daily_ret'][i] >0) :
        signal.append(1)
    elif (joined_df['Bear_sum'][i] >0.0 and joined_df['Bear_sum'][i] <= 400) and joined_df['daily_ret'][i] < 0:
        signal.append(-1)
    else:
        signal.append(0)
```

```python
joined_df['signal'] = signal
```

```python
joined_df
```

```python
joined_df=joined_df.dropna()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
y = joined_df['signal']
X = joined_df.drop(columns=['close','signal'])
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train.shape
# Scale the data
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
lr.fit(X_train_scaled, y_train)
```

```python
# Make predictions using the test data
y_pred = lr.predict(X_test_scaled)

results = pd.DataFrame({
    "Prediction": y_pred, 
    "Actual": y_test
}).reset_index(drop=True)

results.head()
```

```python
# Import Modules
from sklearn.metrics import confusion_matrix,classification_report
```

```python
# Print confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# Print classification report
print(classification_report(y_test, y_pred))
```
